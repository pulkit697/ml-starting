```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
#default is mean
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

# Count plots
 we don't need to give y here, y is already fixed, that is count

```python
sns.countplot(x='time',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

# Box Plot
It contains minimum, maximum and average values. Box signifies where most value lies, and mark outliers as seperate points.

```python
sns.boxplot(x = 'day', y='total_bill', data=tips)
```

```python
sns.boxplot(x = 'day', y='total_bill', data=tips, hue='sex')
# hue divides each categorical data into sub parts
```

# violin graph
width signifies count
inner black line => min, mean and max

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex',split=True)
```

# dist Plot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],bins=30)
```

# Kde Plot

```python
sns.kdeplot(tips['total_bill'])
```

# Joint Graph

```python
sns.jointplot(x = 'total_bill',y='tip',data=tips)
# kind = 'scatter' by default
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=tips, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=tips, kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=tips, kind = 'reg')
```

# Pair plot

```python
sns.pairplot(data=tips)
```

```python
sns.pairplot(data=tips,hue='sex')
```

# Heat maps

```python
tips_corr = tips.corr()
```

```python
tips_corr
```

```python
sns.heatmap(tips_corr)
```

```python
#printing values in heat map
sns.heatmap(tips_corr,annot=True)
```

Loading another data set:

```python
flights = sns.load_dataset("flights")
```

```python
flights
```

```python
# pivot table in pandas is used to classify column values with that of other
# index => row
# columns => col
# values => the values
flights_pivot = flights.pivot_table(index='month',columns='year',values='passengers')
flights_pivot
```

```python
sns.heatmap(flights_pivot)
```

```python
sns.heatmap(flights_pivot,square=True)
```

```python
sns.heatmap(flights_pivot,cmap='coolwarm')
```
