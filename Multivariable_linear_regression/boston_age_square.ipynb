```python
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
import pandas as pd
import matplotlib.pyplot as plt
```

```python
boston = datasets.load_boston()
x = boston.data
y = boston.target
```

```python
# we are creating a new dataset which also includes dependency on age^2
df = pd.DataFrame(x)
df.columns = boston.feature_names
df["AGE_AGE"] = df['AGE']**2
```

```python
x2 = df.values
```

```python
print(x.shape)
print(x2.shape)
```

```python
alg1=LinearRegression()
alg2=LinearRegression()
```

```python
x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,random_state = 0)
x2_train,x2_test,y2_train,y2_test = model_selection.train_test_split(x2,y,random_state = 0)
```

```python
alg1.fit(x_train,y_train)
alg2.fit(x2_train,y2_train)
```

```python
y_pred = alg1.predict(x_test)
y2_pred = alg2.predict(x2_test)
print("Score for dataset 1: ",alg1.score(x_train,y_train))
print("Score for dataset 2: ",alg2.score(x2_train,y2_train))
```

```python
print(x_test.shape,"    ",y_test.shape)
print(x_test.shape,"    ",y_pred.shape)
print(x2_test.shape,"    ",y2_pred.shape)
```

```python
# plot
plt.scatter(x_test[:,0],y_test,color='blue',label='original data')
plt.scatter(x_test[:,0],y_pred,color='red',label='single age')
plt.scatter(x2_test[:,0],y2_pred,color='green',label='age square')
plt.legend
plt.show()
```
