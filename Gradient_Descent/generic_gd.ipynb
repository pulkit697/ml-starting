```python
import numpy as np
```

```python
def cost(x,y,m):
    N = len(x)
    M = len(x[0])
    total=0
    for i in range(N):
        temp=y[i]
        for k in range(M):
            temp-=m[k]*x[i,k]
        temp = temp**2
        total+=(1/N)*temp
    return total
```

```python
def take_one_step(x,y,m,alpha):
    N = len(x)
    M = len(x[0])
    print(m)
    slope = np.zeros((M,))
    for j in range(M):
        for i in range(N):
            temp=y[i]
            for k in range(M):
                temp-=m[k]*x[i,k]
            temp*=x[i,j]
            temp*=(-2/N)
            slope[j]+=temp
    for j in range(M):
        m[j]-=alpha*slope[j]
    return m
```

```python
def gd(x,y,alpha,iterations):
    M = len(x[0])
    m = np.zeros((M,))
    for i in range(iterations):
        print(m)
        m=take_one_step(x,y,m,alpha)
        print(cost(x,y,m))
    return m
```

```python
def run():
    data = np.loadtxt("data.csv",delimiter=',')
    alpha = 0.001
    iterations = 100
    x = data[:,0]
    y = data[:,1]
    N = len(x)
    z = np.ones((N,))
    x = np.c_[x,z]
#     print(x)
#     print(x[0,0])
#     print(x[0,1])
#     print(x[50,0])
#     print(x[50,1])
    gd(x,y,alpha,iterations)
```
