```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-1.5,1.6,0.1)
y2 = x**3
y1 = x**2
plt.plot(x,y1,color='green',linewidth='1.5',label='x^2',)
plt.plot(x,y2,color='red',label='x^3')
plt.legend()
plt.text(0,-0.35,"(0,0)",fontsize='12',fontstyle='oblique')
plt.scatter([0],[0],color='black')
# plt.axis([-1.6,1.6,-4,4])
plt.ylabel('y')
plt.xlabel('x')
plt.grid()
plt.title('x2 vs x3')
plt.show()
```

```python
# pie graphs
sizes = [3,2,4,1]
colors=['red','purple','pink','green']
labels=["A",'B','C','D']
explode=[0,0.10,0,0]
plt.pie(sizes,labels=labels,counterclock=False,startangle=180,colors=colors,autopct="%1.2f%%",explode = explode,shadow=True,
        wedgeprops={
            "edgecolor":'black',
            'linewidth':1
        })
plt.axis('equal')
plt.title('Pie chart')
plt.show()
```
