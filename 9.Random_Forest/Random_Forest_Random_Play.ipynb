# Random Forest

# Random Forest Random Play

Fixing max depth gave better results

```python
clf3 = RandomForestClassifier(random_state=0,max_depth=7)
clf3.fit(x_train,y_train)
print(clf3.score(x_train,y_train),"     ",clf3.score(x_test,y_test))
# Huge Improvement
```

changing number of trees in the forest:

```python
clf4 = RandomForestClassifier(random_state=0,n_estimators=513)
clf4.fit(x_train,y_train)
print(clf4.score(x_train,y_train),"     ",clf4.score(x_test,y_test))
# A little improvement
```

```python
clf5 = RandomForestClassifier(random_state=0,min_samples_split=20)
clf5.fit(x_train,y_train)
print(clf5.score(x_train,y_train),"     ",clf5.score(x_test,y_test))
# Huge Improvement
```

```python
clf6 = RandomForestClassifier(random_state=0,max_features="sqrt")
clf6.fit(x_train,y_train)
print(clf6.score(x_train,y_train),"     ",clf6.score(x_test,y_test))
# No change
```

```python
clf7 = RandomForestClassifier(random_state=0,bootstrap=False)
clf7.fit(x_train,y_train)
print(clf7.score(x_train,y_train),"     ",clf7.score(x_test,y_test))
# decrement
```

```python
clf8 = RandomForestClassifier(random_state=0,max_depth=7,min_samples_split=20)
clf8.fit(x_train,y_train)
print(clf8.score(x_train,y_train),"     ",clf8.score(x_test,y_test))
# Nice!
```
